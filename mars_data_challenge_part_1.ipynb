```python
# Import pandas, Splinter, BeautifulSoup, and ChromeDriverManager
import pandas as pd
from bs4 import BeautifulSoup as soup
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the Mars News site
url = 'https://redplanetscience.com/'
browser.visit(url)
```

```python
# Set up HTML Parser
html = browser.html
news_soup = soup(html, 'html.parser')
```

```python
# Create an empty ditionary and an empty list
articles = []

# Name title and preview for parsing
titles = news_soup.find_all('div', class_='content_title')
previews = news_soup.find_all('div', class_='article_teaser_body')
```

```python
# Use for loop to iterate through each article for title and preview pairs
for title in titles:
    article_data = {}
    print('title:' +title.text)
    article_data['title'] = title.text

    for preview in previews:
        article_date = {}
        print('preview:' +preview.text)
        article_data['preview'] = preview.text
        
        articles.append(article_data)
```

```python
# Store article titles and previews from list into json file with json.dumps
import json
mars_articles = json.dumps(articles)
print(mars_articles)
```
